```python
import pandas as pd
import numpy as np
```

```python
num = pd.read_csv("./data/2017q4/num.txt", delimiter= "\t")
pre = pd.read_csv("./data/2017q4/pre.txt", delimiter= "\t")
```

```python
num['version'].value_counts()
```

```python
random_adsh = num['adsh'].unique()[12]
random_co = num[(num['adsh']==random_adsh)]
random_co.head()
random_co_bs_tags = pre[(pre['adsh']==random_adsh) & (pre['stmt']=='BS')].tag
bs = random_co[(random_co.tag.isin(random_co_bs_tags))]
bs = bs.sort_values(by=['ddate']).drop_duplicates(subset="tag", keep="last")
bs
```
